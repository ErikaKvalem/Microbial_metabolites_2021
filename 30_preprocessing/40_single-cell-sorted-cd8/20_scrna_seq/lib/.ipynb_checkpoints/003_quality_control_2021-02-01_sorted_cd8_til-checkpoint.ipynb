```python
import scanpy as sc
import seaborn as sns
```

```python
adata_path = "/data/projects/2021/MicrobialMetabolites/single-cell-sorted-cd8/2019-10-29_sorted_cd8/analyses_icbi/scrna_seq_data/2019-10-29_sorted_cd8_merged_data.h5ad"
```

```python
adata = sc.read_h5ad(adata_path)
```

```python
# mitochondrial genes
adata.var["mt"] = adata.var_names.str.startswith("mt-")
```

```python
sc.pp.calculate_qc_metrics(
    adata, qc_vars=["mt"], inplace=True, percent_top=[20], log1p=True)
adata
```

calculate_qc_metrics

percent_top : Optional, optional (default: (50, 100, 200, 500))
    Which proportions of top genes to cover. If empty or `None` don't
    calculate. Values are considered 1-indexed, `percent_top=[50]` finds
    cumulative proportion to the 50th most expressed gene.
    
log1p : bool, optional (default: True)
Set to `False` to skip computing `log1p` transformed annotations.
